```python
df=pd.read_csv('GroceryStoreDataSet.csv')
```

```python
df
```

```python
data=[]
for i in range(0,len(df)):
    data.append([str(df.values[i,j])for j in range(0,5)])
        
```

```python
data
```

```python
from apyori import apriori

# Training Apriori on the dataset
rules = apriori(data, min_support = 0.08, min_confidence = 0.7, min_lift = 3, min_length = 2,
               max_length = 3)

# Create list of results
results = list(rules)
```

```python
results
```

```python
def inspect(results):
    li          = [tuple(result[2][0][0])[0] for result in results]
    ri          = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(li, ri, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Item', 'Right Item', 'Support', 'Confidence', 'Lift'])
```
